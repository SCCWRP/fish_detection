```python
# imports
import cv2
import skimage
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
# opencv uses BGR whereas the other libraries use RGB
# so if i am going to display an image, if i read it in with cv2, i should convert it to RGB
fish4 = cv2.imread('test_images/fish4.png')
fish4 = cv2.cvtColor(fish4, cv2.COLOR_BGR2RGB)
plt.imshow(fish4)
```

```python
grayfish = cv2.cvtColor(fish4, cv2.COLOR_RGB2GRAY)
plt.imshow(grayfish)
```

```python
ret, thresh = cv2.threshold(grayfish, 150, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thresh)
```

```python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours = [c for c in contours if (200 < len(c) < 2000)]
```

```python
fish_w_contours = cv2.drawContours(fish4, contours, -1, (0, 255, 0), 2)
plt.imshow(fish_w_contours)
```
